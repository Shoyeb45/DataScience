```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = list(range(1, 11))
y = list(range(2, 21, 2))
```

```python
plt.plot(x, y, marker = 'o', linestyle='-', color = 'b', label='data')
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.title("Plot")
plt.show()
```

```python
# Figure
fig = plt.figure(figsize = (4, 4), facecolor="gray")
fig.suptitle("Figure")
ax = fig.add_subplot()
ax.set_title("Hii", loc="right", color= "White")
# ax.plot(x, y, label="shoyeb")
ax.grid()
ax.legend()
# ax.xaxis.set_major_formatter()
ax.scatter(x,y)
plt.show()
```

```python
d = [1, 2, 3, 4]
r = [i**2 for i in d]
plt.plot(d, r, 'ro--', color='blue', markerfacecolor="black", marker='o')
plt.show()
```

```python
x = np.arange(1, 10, 1)
y = x ** 2
z = x ** 3
fig, ax = plt.subplots(1, 1)
ax.plot(x, y, label = 'quadratic')
ax.plot(x, z, label = 'Cubic', color = "black")
ax.spines['right'].set_visible(False)
ax.legend()
ax.spines['top'].set_color('none')
ax.set_xlabel("x")
ax.set_ylabel("$x^2$")
ax.set_title("Graph of $x^2$")

plt.show()
```
