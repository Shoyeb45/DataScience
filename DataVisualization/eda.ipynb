```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
# histogram
plt.figure(figsize=(10, 6))
sns.histplot(df["sepal_length"], bins = 30)
plt.title("Histogram of Sepal Length")
plt.show()
```

```python
# density plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df["sepal_length"])
plt.title("Density plot of sepal length")
plt.show()
```

```python
# Rug plot
plt.figure(figsize = (10, 6))
sns.kdeplot(df["sepal_length"])
sns.rugplot(df["sepal_length"])
plt.title("Rug plot of Sepal Length")
plt.show()
```

```python
# Box Plot
plt.figure(figsize = (10, 6))
sns.boxplot(x = "species", y = "sepal_length", data = df)
plt.title("Box plot of Sepal Length")
plt.show()
```

```python
# Correlation plot
plt.figure(figsize = (10 ,6))
sns.heatmap(df.iloc[:, 0:4].corr(), annot = True )
plt.title("Correlation plot of Iris Dataset")
plt.show()
```
