## Importing Libraries

````python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
````

## Creating and Manipulating Networks

````python
# Creating a Graph
G = nx.Graph() # Creates an empty graph
# Adding Nodes
G.add_nodes_from([2, 3, 4])  # Adding multiple nodes
# Adding Edges
G.add_edges_from([(2, 3), (3, 4), (4, 1)])
G.add_edge(1, 2)
````

### Getting Node and Edge Lists

````python
print('Nodes:', G.nodes())
print('Edges:', G.edges())
````

### Saving and Loading Networks

````python
nx.write_edgelist(G, 'network.edgelist')  # Save to file
G_loaded = nx.read_edgelist('network.edgelist')  # Load from file
````

## Visualizing the Network

````python
nx.draw(G, with_labels=True)
plt.show()
````

## Different Types of Networks

### Regular Network
- A regular network is a graph where each node has the same number of edges.

````python
G_reg = nx.random_regular_graph(d=3, n=20)
nx.draw(G_reg, with_labels=True, node_color='lightblue')
plt.title('Regular Network')
plt.show()
````

### Random Network
- A random network is created by connecting nodes randomly based on a probability.

````python
G_rand = nx.erdos_renyi_graph(n=20, p=0.2)
nx.draw(G_rand, with_labels=True, node_color='lightgreen')
plt.title('Random Network')
plt.show()
````

### Scale-Free Network
- A scale-free network follows a power-law distribution where a few nodes have many connections.

````python
G_sf = nx.barabasi_albert_graph(n=20, m=2)
nx.draw(G_sf, with_labels=True, node_color='lightcoral')
plt.title("Scale-Free Network")
plt.show()
````

### Small-World Network
- A small-world network has high clustering and short path lengths, resembling real-world social networks.

````python
G_sw = nx.watts_strogatz_graph(n=20, k=4, p=0.3)
nx.draw(G_sw, with_labels=True, node_color='lightpink')
plt.title('Small-World Network')
plt.show()
````

## Network Metrics and Analysis

### Clustering Coefficient
-  The clustering coefficient measures the tendency of nodes to form tightly-knit groups or triangles. Higher clustering indicates a stronger tendency for nodes to be interconnected

````python
clustering_coeff = nx.clustering(G)
print('Clustering Coefficients:', clustering_coeff)
````

### Path Finding
- Finding paths in a network is essential for understanding connectivity. The shortest path between two nodes represents the minimum steps required to travel between them.

````python
shortest_path = nx.shortest_path(G, source=1, target=3)
print('Shortest path from 1 to 3:', shortest_path)
````

### Average Path Length
- The average path length provides an idea of how "close" nodes are in a network. Smaller values indicate that nodes are well-connected.

````python
if nx.is_connected(G):
    avg_path_length = nx.average_shortest_path_length(G)
    print('Average Path Length:', avg_path_length)
````

## Network Visualization and Distribution Plots

### Degree Distribution Plot
- The degree distribution helps analyze how connections are spread across the network. Many social networks follow a power-law degree distribution.

````python
degrees = [G.degree(n) for n in G.nodes()]
sns.histplot(degrees, kde=True)
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution')
plt.show()
````

### Row Plot
- A simple row plot helps visualize node degrees in a sequential order.

````python
plt.plot(degrees, marker='o')
plt.xlabel('Node')
plt.ylabel('Degree')
plt.title('Degree Row Plot')
plt.show()
````

### Log-Log Plot
- A log-log plot is useful for identifying scale-free network properties where few nodes have high degrees, while most nodes have a low degree.

````python
plt.loglog(degrees, marker='o')
plt.xlabel('Log(Node)')
plt.ylabel('Log(Degree)')
plt.title('Log-Log Degree Distribution')
plt.show()
````
