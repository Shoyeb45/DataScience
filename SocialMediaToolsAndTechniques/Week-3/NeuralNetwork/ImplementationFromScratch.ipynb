# Implementation of neural network from scratch using numpy

```python
import numpy as np
```

```python
data = np.array([[1, 2, 3],    # Data set for neural network
                 [1, 0, 2],
                 [0, 1, 4],
                 [2, 1, 4]])

# In each row we  have samples, X1, X2,..
```

- We'll build simplets neural network, like this one:



```python
# Initialize weight matrix
np.random.seed(1)
W = np.random.random((3, 1))
# W = np.array([[[1], [0], [1]]])
```

```python
W
```

```python
# Perform forward pass

def forward_pass(input_layer, weight_matrix):
    for iterations in range(1):
        output_matrix = np.dot(input_layer, weight_matrix) # This is X^T.W  - Perceptron
        # print(output_matrix)
        # Pass it in sigmoid function
        output_layer = 1 / (1 + np.exp(-output_matrix))
    return output_layer
```

```python
print(f"Input: \n{data}")
print(f"Output: \n{forward_pass(data, W)}")
```

## Now let's change the number of output nodes



```python
# Now we need to build 3*2 size of weight matrix
W = np.random.random((3, 2))
W
```

```python
print(f"Input: \n{data}")
print(f"Output: \n{forward_pass(data, W)}")
```

## Now let's create hidden layer and build the network, like this:



```python
# Change weight matrix
W = np.random.random((3, 4))
W
```

```python
# Define another weight matrix V for hidden layer and output layer
V = np.random.random((4, 2))
V
```

```python
# Perform forward pass

for iterations in range(1):
    input_layer = data
    
    # X^T.W
    hidden_layer = np.dot(data, W)
    
    # Pass in sigmoid function
    hidden_layer = 1 / (1 + np.exp(-hidden_layer))
    
    # Output layer
    output_layer = np.dot(hidden_layer, V)
    output_layer = 1 / (1 + np.exp(-output_layer))
    
```

```python
print(f"Input: \n{data}")
print(f"Output: \n{output_layer}")
```
