```python
import pandas as pd
# Series
s1=pd.Series(['A','B','C'])
s2=pd.Series(['D','E','F'])
```

```python
s1.index = [1, 2, 3]
s2.index = [1, 2, 3]
```

```python
#concat function

s = pd.concat([s1, s2])
s
```

```python
# dataframe example

df1 = pd.DataFrame(
    { "Name": ["Braund, Mr. Owen Harris","Allen, Mr. William Henry","Bonnell, Miss. Elizabeth"],
     "Age": [22, 35, 58],
     "Gender": ["male", "male", "female"]
})

df1.index = [1, 2, 3]

df2 = pd.DataFrame(
    {
         "Name": ["Ravi","Mr Henry","Elizabeth"],
         "Age": [24, 31, 68],
         "Gender": ["male", "male", "female"],
    }
)

df2.index = [4, 5, 6]
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], axis = 1)
```

```python
df1 = pd.DataFrame({'id': ['A01', 'A02', 'A03', 'A04'],
					'Name': ['ABC', 'PQR', 'DEF', 'GHI']})

df2 = pd.DataFrame({'id': ['A01', 'A02', 'A03', 'A04'],
					'Name': ['ABC', 'PQR', 'DEF', 'GHI']})
c = pd.concat([df1, df2], verify_integrity=False)
```

```python
c
```

```python
# pd.concat([df1, df2], verify_integrity=True), this will give error
```

```python
# ignore index
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
# adding index keys

df3 = pd.DataFrame(  {
    "id" : ["A05", "A06", "A07", "A08"],
    "Name" : ['ABC', 'EQR', 'DRF', 'GUI']
})

pd.concat([df1, df2, df3], keys=["x", "y" , "z"])
```
