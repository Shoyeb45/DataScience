```python
import pandas as pd

df = pd.read_excel("E:/Notes/DataScience/Pandas/SampleSuperstore.xlsx")
```

```python
df.columns
```

```python
# 1. average sales amount per order
df.groupby("Order ID")["Sales"].mean()
```

```python
# 2.Determine the number of unique customer IDs

len(df["Customer Name"].unique())
```

```python
# 3. top 5 states have the highest total sales, Aggregate sales by state and rank them to find the top 5.

l = df.groupby('State')["Sales"].sum()
```

```python
l = l.sort_values(ascending=False)
print(f"Top 5 states\n{l.head(5)}")
```

```python
# 4.  What is the distribution of sales across different shipping modes?
```

```python
df.groupby("Ship Mode")["Sales"].mean()
```

```python
df.groupby("Ship Mode")["Sales"].max()
```

```python
df.groupby("Ship Mode")["Sales"].min()
```

```python
df.groupby("Ship Mode")["Sales"].describe()
```

```python
# 5. Which product sub-category has the lowest average profit?
```

```python
s = df.groupby("Sub-Category")["Profit"].mean()
```

```python
print(f"Lowest average profit :\n {s.idxmin()}")
```

```python
# 6. How does the discount level affect the average order quantity?
```

```python
pd.crosstab(df["Quantity"], df["Discount"])
```

```python
pd.crosstab(df["Quantity"].mean(), df["Discount"])
```

```python
q = df.groupby(by = "Quantity")
for group, dataFrame in q:
    print(group)
    print(dataFrame)
```

```python
# 7. What are the top 3 most profitable products?
```

```python
l = df.groupby('Product Name')["Profit"].sum()
l = l.sort_values(ascending=False)
print(f"Top 3 product by profit\n{l.head(3)}")
```

```python
# 8. Which month of the year has the highest total sales?
```

```python
x = df.groupby("Order Date")["Sales"].sum()
```

```python
date = x.idxmax()
print(f"Month of the highest total sales:{date.month}")
```

```python
# 9.  What is the profit ratio (profit/sales) for each category, and which category has the highest ratio?
```

```python
profit = df.groupby("Category")["Profit"].sum()
```

```python
sales = df.groupby("Category")["Sales"].sum()
```

```python
ratio = profit/sales
```

```python
ratio
```

```python
ratio.idxmax()
```
