```python
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
import sklearn as sk
```

```python
# creating dataframe
df = pd.DataFrame ( {
    'A' : [1, 2, np.nan, 4, 5],
    'B' : [np.nan, 2, 3, 4, 5],
    'C' : [1, 2, 3, np.nan, np.nan]
})
```

```python
df
```

```python
imputer = SimpleImputer() 
```

```python
# Now we have to fill the NaN values with mean of the remaining data
# First creating instance of imputer with filling(strategy) value as mean
impute = SimpleImputer(missing_values = np.nan, strategy = "mean")
```

```python
# Now creating new data frame to with added missing values
df_imputed_array = imputer.fit_transform(df)
df_imputed = pd.DataFrame(df_imputed_array, columns=df.columns)
print(df_imputed)
```
