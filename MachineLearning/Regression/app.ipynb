```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("E:/Notes/DataScience/MachineLearning/Regression/car_price.csv")
```

```python
df.head()
```

```python
df.groupby("make").agg("selling_price").aggregate("mean")
```

```python
df.loc[:, "make"] = df.groupby(["make"])["selling_price"].transform("mean")  # replacing values of 'make' with the average values of corresponding model
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df["selling_price"]
df.drop(columns=["selling_price"],inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.2, random_state=10)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
